```python
import re
import json
import random
from pathlib import Path
from time import perf_counter


# ----------------------------
# REGEX
# ----------------------------
WORD_RE = re.compile(r"[A-Za-zÀ-ÖØ-öø-ÿ']+")

# ----------------------------
# STOPWORDS
# ----------------------------
STOPWORDS = {
    "en": {"the","and","to","of","a","in","that","is","it","for","on","with","as","this",
           "be","are","you","your","can","will","not","or","if","we","from","by","at","an",
           "have","has","do","does","what","which","when","where","how","why","about","would",
           "could","should","there","their","they","them","these","those","one","more","use",
           "using","make","write","explain","please","help"},
    "de": {"der","die","das","und","ist","ich","du","sie","er","es","wir","nicht","ein","eine",
           "zu","mit","auf","für","von","wie","was","warum","wenn","dass","kann","sind","haben",
           "werden","auch","oder","aber","bitte","über","hallo"},
    "nl": {"de","het","een","en","van","ik","je","niet","dat","dit","die","op","voor","met",
           "als","zijn","hebben","worden","wat","waar","waarom","hoe"},
    "vi": {"và","là","của","có","cho","trong","một","không","tôi","bạn","hãy","này","để"},
    "es": {"el","la","los","las","y","de","que","en","un","una","es","por","para","con","como","no","se"},
    "fr": {"le","la","les","et","de","des","du","un","une","est","dans","pour","avec","que","qui"},
    "pt": {"o","a","os","as","e","de","que","em","um","uma","é","para","com","como","não"},
    "it": {"il","lo","la","gli","le","e","di","che","in","un","una","è","per","con"},
}

# ----------------------------
# FIRST USER PROMPT ONLY
# ----------------------------
def get_first_user_prompt(tree):
    conversations = tree.get("conversations", [])

    for message in conversations:
        if message.get("from") == "human":
            return message.get("value", "")

    return ""


# ----------------------------
# TEXT NORMALIZATION (prompt-only)
# ----------------------------
def normalize_words(text):
    return [w.lower().strip("'") for w in WORD_RE.findall(text)]


def count_words(text):
    return len(normalize_words(text))


def has_question(text):
    return "?" in text


# ----------------------------
# LANGUAGE SCORING (prompt only)
# ----------------------------
def count_stopwords(words):
    return {
        lang: sum(w in stopwords for w in words)
        for lang, stopwords in STOPWORDS.items()
    }


def get_language_scores_from_prompt(prompt):
    words = normalize_words(prompt)

    stopword_scores = count_stopwords(words)

    english_score = stopword_scores["en"]
    strongest_other = max(v for k, v in stopword_scores.items() if k != "en")

    return {
        "words": words,
        "word_count": len(words),
        "stopword_scores": stopword_scores,
        "english_score": english_score,
        "strongest_other": strongest_other
    }


# ----------------------------
# LANGUAGE FILTER
# ----------------------------
def is_english_tree(tree,
                     min_words=5,
                     min_english_stopwords=2,
                     min_margin=1):

    prompt = get_first_user_prompt(tree)

    # ONLY FIRST PROMPT
    scores = get_language_scores_from_prompt(prompt)

    if scores["word_count"] < min_words:
        return False

    stopword_scores = scores["stopword_scores"]

    english_score = stopword_scores["en"]
    strongest_other = max(v for k, v in stopword_scores.items() if k != "en")

    return (
        english_score >= min_english_stopwords
        and (english_score - strongest_other) >= min_margin
    )


# ----------------------------
# FILTER DATA
# ----------------------------
def filter_english_trees(data):
    return [tree for tree in data if is_english_tree(tree)]


# ----------------------------
# RANDOM SAMPLE VIEW
# ----------------------------
def show_random_questions(data, n=10, max_chars=700):
    for i in random.sample(range(len(data)), min(n, len(data))):
        tree = data[i]

        prompt = get_first_user_prompt(tree)

        print("=" * 100)
        print(f"Index: {i}")
        print(prompt[:max_chars])


# ----------------------------
# LOAD DATA
# ----------------------------

def load_json(path):
    with Path(path).open("r", encoding="utf-8") as file:
        return json.load(file)
    
part1 = load_json("C:/Users/heike/Desktop/Stackfuel/Portfolio/llm-sustainability-analysis/01_data/01_raw/messy_dataset/sg_90k_part1.json")

part2 = load_json("C:/Users/heike/Desktop/Stackfuel/Portfolio/llm-sustainability-analysis/01_data/01_raw/messy_dataset/sg_90k_part2.json")

data = part1 + part2

# ----------------------------
# RUN FILTER
# ----------------------------
start = perf_counter()

english_data = filter_english_trees(data)

duration = perf_counter() - start

# ----------------------------
# SAVE
# ----------------------------

def save_json(data, path):
    path = Path(path)
    path.parent.mkdir(parents=True, exist_ok=True)

    with path.open("w", encoding="utf-8") as file:
        json.dump(data, file, ensure_ascii=False, indent=2)

save_json(
    english_data,
    "C:/Users/heike/Desktop/Stackfuel/Portfolio/llm-sustainability-analysis/01_data/02_processed/sharegpt_english.json"
)

print(f"Original: {len(data)}")
print(f"English only: {len(english_data)}")
print(f"Removed: {len(data) - len(english_data)}")
print(f"Time: {duration:.2f}s")
print(f"Trees/sec: {len(data)/duration:.0f}")

show_random_questions(english_data, n=10)
```
